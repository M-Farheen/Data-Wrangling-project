```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
#Read Data using panda
df = pd.read_csv('fifa21 raw data v2.csv',low_memory = False)
```

```python
#check data 
df.shape
```

```python
#check the Null Data
df.isnull()
```

```python
#Remove Duplicates
df.drop_duplicates(inplace = True)
```

```python
#Drop the columns that we don't Need
df.drop(['ID','Name', 'photoUrl', 'playerUrl','Loan Date End'], axis = 1, inplace = True)
```

```python
#Rename the Columns
df.rename(columns={'â†“OVA': 'OVA', 'LongName': 'Name'}, inplace=True)
```

```python
#cleans the column "club"
df.Club = df.Club.str.strip('\n\n\n\n')
```

```python
#check the contract column
df.Contract.unique()
```

```python
#change contract column 
def Contract_val(a):
    if "On Loan" in a:
        a = 'On Loan'
        return a
    elif '~' in a:
        a = 'Active'
        return a
    elif 'Free' in a:
        a = "Free"
        return a
```

```python
df.Contract = df.Contract.apply(Contract_val)
```

```python
df['Height'].unique()
```

```python
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"","")
        inch = int(x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
    elif x[-1] == "m":
        return int(x[:-2])
```

```python
df['Height'] = df['Height'].apply(convert_height)
```

```python
df['Weight'].unique()
```

```python
def convert_weight(x):
    if 'lbs' in x:
        weight_lbs = float(x.strip('lbs'))
        weight_kg = int(weight_lbs * 0.45359237)
        return str(weight_kg)
    else:
        return int(x[:-2])
```

```python
df['Weight'] = df['Weight'].apply(convert_weight)
```

```python
#changes the column Preferred Foot
print(df["Preferred Foot"].unique())
df["Preferred Foot"] = np.where(df["Preferred Foot"]=="Right", "R", "L")
```

```python
#Rename the Columns
df.rename(columns={'Height': 'Height(cm)', 'Weight(cm)': 'Weight(kg)'}, inplace=True)
```

```python
df['Joined'] = pd.to_datetime(df['Joined'])
```

```python
df.Value.unique()
```

```python
def Value_to_numeric(x):
    if 'K' in x:
        x = x.replace('â‚¬','').replace('K','')
        x = float(x) * 1000
        return int(x)
    else:
        x = x.replace('â‚¬','').replace('M','')
        x = float(x) * 1000000
        return int(x)

df['Value'] = df['Value'].apply(Value_to_numeric)
```

```python
df.Wage.unique()
```

```python
def wage_to_numeric(x):
    if 'K' in x:
        x = x.replace('â‚¬','').replace('K','')
        x = float(x) * 1000
        return int(x)
    else:
        x = x.replace('â‚¬','').replace('M','')
        x = float(x) * 1000000
        return int(x)

df['Wage'] = df['Wage'].apply(wage_to_numeric)
```

```python
df['Release Clause'].unique()
```

```python
def Release_to_numeric(x):
    if 'K' in x:
        x = x.replace('â‚¬','').replace('K','')
        x = float(x) * 1000
        return int(x)
    else:
        x = x.replace('â‚¬','').replace('M','')
        x = float(x) * 1000000
        return int(x)

df['Release Clause'] = df['Release Clause'].apply(Release_to_numeric)
```

```python
df[['W/F', 'SM', 'IR']] = df[['W/F', 'SM', 'IR']].replace('â˜…', '', regex=True).astype(int)
df["Age"] = df["Age"].astype(int)
```

```python
df.Hits.unique()
```

```python
def hitchange(h):
    if isinstance(h, float):
        return h
    elif "K" in h:
        h = h.replace('K', "")
        h = float(h) * 1000
        return round(h)
    else:
        return h
df['Hits'] = df['Hits'].apply(hitchange)
```

```python
df['Hits'] = df['Hits'].fillna(0)
```
